```python
is_ = data['signup_time']==data['purchase_time']
```

```python
df=is_
df.head()
```

```python
data=pd.concat([data,df],axis=1)
```

```python
data.head(5)
```

```python
data[data.signup_time==data.purchase_time]
```

```python
data = pd.get_dummies(data, columns=['source','browser','sex',0])
```

```python
data.info()
```

```python
data.head(5)
```

```python
data=data.drop(['age','user_id','signup_time','purchase_time','purchase_value','device_id','0_False'],axis=1)
```

```python
data.head(5)
```

```python
x = data.drop("class",axis=1)
y = data['class']
```

```python
print (x.shape, y.shape)
```

```python
#import model specific libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=100)
```

```python
#validate the shape of train and test dataset
print (X_train.shape)
print (y_train.shape)

print (X_test.shape)
print (y_test.shape)
```

```python
logreg =LogisticRegression(class_weight='balanced')
```

```python
sd=logreg.fit(X_train,y_train)
```

```python
logreg.score(X_train,y_train)
```

```python
predictions=logreg.predict(X_test)
```

```python
print(predictions)
```
